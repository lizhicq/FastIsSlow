```python
import pandas as pd 
# Re-import the file

xl2 = pd.ExcelFile('data/附件1-葡萄酒品尝评分表.xls')


import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from sklearn.metrics import mean_absolute_error
from scipy.stats import t, sem
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math

# Load the data
group1_white_wine_scores = pd.read_excel('/path/to/第一组白葡萄酒品尝评分.xls', header=None)
group2_white_wine_scores = pd.read_excel('/path/to/第二组白葡萄酒品尝评分.xls', header=None)

# Define a function to preprocess the wine evaluation data
def preprocess_wine_evaluation_data(df):
    # Initialize an empty DataFrame to hold the preprocessed data
    preprocessed_data = pd.DataFrame()

    # Initialize the sample name
    sample_name = None

    # Iterate over the rows in the DataFrame
    for index, row in df.iterrows():
        # Check if the row contains a new sample name
        if "酒样品" in str(row[2]):
            sample_name = row[2]
            continue
        
        # Check if the row contains a new category name
        if pd.notnull(row[1]) and pd.isna(row[2]):
            continue

        # Check if the row contains scores
        if pd.isna(row[0]) and pd.notnull(row[2]) and pd.notnull(row[3]):
            # Construct a new row for the preprocessed data
            new_row = {"酒样": sample_name, "评价标准": row[2]}
            for i in range(3, len(row)):
                new_row["品酒员" + str(i-2) + "号"] = row[i]
            
            # Append the new row to the preprocessed data
            preprocessed_data = preprocessed_data.append(new_row, ignore_index=True)

    # Return the preprocessed data
    return preprocessed_data

# Preprocess the white wine evaluation data
group1_white_wine_scores_preprocessed = preprocess_wine_evaluation_data(group1_white_wine_scores)
group2_white_wine_scores_preprocessed = preprocess_wine_evaluation_data(group2_white_wine_scores)

# Save the preprocessed data to CSV files
group1_white_wine_scores_preprocessed.to_csv('/path/to/group1_white_wine_scores_preprocessed.csv', index=False)
group2_white_wine_scores_preprocessed.to_csv('/path/to/group2_white_wine_scores_preprocessed.csv', index=False)





```
