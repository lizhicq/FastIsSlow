```python
# Import the modules
import os
import openai

# Set the API key
openai.api_key = "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"
# Set the directory path
path = "/Users/lizhicq/Desktop/PSC"
# Get the list of files
files = os.listdir(path)

# Loop through the files
for file in files:
    # Get the full file path
    file_path = os.path.join(path, file)
    # Open the file
    with open(file_path, "r") as f:
        # Read the file contents
        data = f.read()
        # Create a prompt for the OpenAI API
        prompt = f'''
        This is the performance data of {file}:\n{data}\n\nPlease help me write a peer review for the person. Requirement:
        1. Two parts: strengths and weaknesses 
        2. please write each part with bullet points, starting with good or bad performing area, use concrete example or cases to support
        '''
        # Call the OpenAI API
        completion = openai.chat.completions.create(
            model="gpt-4",
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                },
            ],
        )
        # Get the feedback
        feedback = completion.choices[0].message.content
        # Print or save the feedback
        print(f"==========Feedback for {file}=============:\n{feedback}")
```
